# TP PROJ 5 – Méthodes analytiques, Fourier et approximations

Objectif : utiliser les modules d’appui (`Analytical/`, `Fourier/`, `Approx/`) pour valider et comparer les résultats PROJ.


## 1. Formules analytiques Black–Scholes

Dossier : `Analytical/BlackScholes/`.
- `Script_BlackScholes_Prices_Greeks.m` et `Script_BlackScholes_Plot_Greeks_For_Strike.m` : génération de surfaces de prix/grecs.
- `BSM_Greeks.m`, `Price_Exchange_Option_Margrabe_2D.m`, `Price_Spread_Option_BjerksundStensland_2D.m`, `Price_Spread_Option_Kirk_2D.m` : références fermées.
Utilisation :
- produire des tableaux de vérité pour comparer aux méthodes PROJ/Monte Carlo ;
- analyser les sensibilités (vega, gamma) qui serviront pour la calibration.


## 2. Méthodes Fourier

Dossier : `Fourier/` (sous‑dossiers `COS/`, `CarrMadan/`, `CONV/`, `Lewis/`, `HilbertTransform/`, `MellinTransform/`, `Heston/`).
- `Fourier/COS/Price_Exchange_COS2D.m` : extension multi‑actifs du COS.
- Les scripts Carr‑Madan / CONV / Lewis permettent de benchmarker les transformées rapides vs PROJ.
Travail :
- implémenter un pipeline commun (mêmes inputs) pour comparer PROJ vs Carr‑Madan vs COS ;
- tester la reconstruction via transformée de Hilbert ou Mellin pour des payoffs non ordinaires ;
- utiliser `Fourier/Heston/` pour valider les prix de `PROJ/STOCHASTIC_VOL`.


## 3. Approximations fermées

Dossier : `Approx/BlackScholes/Asian/` et `Approx/SABR/`.
- `Price_Asian_VorstApprox_BlackScholes.m` : approximation de Vorst pour asiatiques BSM.
- `SABR_European_AntonovApprox.m`, `SABR_Hagan_Obloj_ImpliedVol.m`, `Script_SABR_European_AntonovApprox.m` : approximations SABR.
Exercices :
- comparer l’approx de Vorst aux prix PROJ/Monte Carlo pour calibrer le nombre de dates ;
- dériver des surfaces d’IV SABR via l’Antonov/Hagan et les confronter aux approches CTMC du TP10 ;
- documenter les domaines de validité (petit tau, beta proche de 1, etc.).
