# TP PROJ 10 – SABR : approximations fermées et CTMC full‑fledged

Objectif : relier les approximations SABR (Hagan/Antonov) aux approches CTMC/PROJ pour pricing d’options (européennes, américaines, barrières).


## 1. Approximations SABR fermées

Dossier : `Approx/SABR/`.
- `SABR_European_AntonovApprox.m`, `SABR_Hagan_Obloj_ImpliedVol.m`, `Script_SABR_European_AntonovApprox.m`.
- Retourne vols implicites et prix analytiques rapides.
Travail :
- générer une surface (strike, maturité) et comparer aux résultats PROJ/CTMC ;
- analyser les limites des approximations (beta, nu, rho extrêmes) ;
- préparer ces sorties comme initial guess pour les solveurs numériques.


## 2. PROJ sous dynamique SABR

Dossier : `PROJ/SABR/European_American_Barrier/` + `Helper_Functions/`.
- `SABR_EurBarAmer_func.m`, `Script_European_Barrier_American.m` : pricers PROJ adaptés à SABR via projection sur la variance stochastique.
- Utilise des approximations de densité (b-splines) analogues aux modules `STOCHASTIC_VOL`.
Exercices :
- comparer le prix PROJ vs approximation Antonov/Hagan sur plusieurs strikes ;
- tester la partie américaine/barrière pour comprendre l’impact de la volatilité stochastique ;
- étudier la sensibilité au nombre de points de grille variance.


## 3. CTMC full‑fledged pour SABR

Ressources : `CTMC/` (fonctions `Q_Matrix.m`, `decorrelate.m`) + scripts SABR.
- Construire une approximation CTMC pour le couple (F, vol) en suivant la même logique que `STOCHASTIC_VOL`.
- Utiliser ensuite les scripts CTMC (barrières, bermudanes) pour obtenir des prix « exacts ».
Travail :
- dériver la matrice génératrice pour SABR avec `beta ≠ 1` et étudier la stabilité ;
- benchmarker CTMC vs PROJ vs approximation fermée sur plusieurs nu/rho ;
- identifier les compromis (taille de grille, coût CPU) d’une implémentation full‑fledged.


## 4. Plan de test et restitution

- Centraliser les comparaisons (tableaux prix/vol implicite) entre approximations fermées, PROJ et CTMC.
- Documenter les réglages recommandés pour chaque approche (taille de grille variance, ordre B‑spline, nombre de points CTMC).
- Préparer un template de rapport synthétisant les écarts et les cas limites (beta proche de 0, maturités longues, strikes extrêmes).
