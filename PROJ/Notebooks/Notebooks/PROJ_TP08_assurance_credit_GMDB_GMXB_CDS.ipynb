# TP PROJ 8 – Produits d’assurance / crédit : GMDB, GMXB, CDS

Concentré sur les produits **assurance‑vie / crédit** présents dans `PROJ/LEVY/` : garanties décès, GMXB avec rachat et credit default swaps.


## 1. Equity‑Linked Death Benefits (ELDB)

Dossier : `PROJ/LEVY/Equity_Linked_Death_Benefits/`.
- `PROJ_GMDB_ComboExpos_Linear.m`, `PROJ_GMDB_PiecewiseConstantMortality_Linear.m` et les scripts `Script_GMDB_ComboExpos.m`, `Script_GMDB_PiecewiseConstantMortality.m`.
- Table de mortalité `lifetable_qx.txt` fournie.
Travail :
- comprendre comment la mortalité est discrétisée (combo exponentielles vs table) ;
- vérifier la projection PROJ conditionnellement à l’événement de décès ;
- expérimenter différentes hypothèses de mortalité et d’exposition aux marchés.


## 2. GMDB avec contributions périodiques (Dollar‑Cost Averaging)

Dossier : `PROJ/LEVY/GMDB_DCA/` + utilitaires `gen_func.m`, `make_combo_2_expos_pmf.m`, `make_mortality_table_pmf.m`.
- `PROJ_GMDB_DCA.m`, `PROJ_GMDB_DCA_Fast.m`, `Script_GMDB_DCA.m` modélisent des contributions régulières et la garantie décès correspondante.
Exercices :
- paramétrer une séquence de versements et visualiser l’évolution de l’actif notionnel ;
- comparer la version « Fast » vs « brute » (gain en performances, limites) ;
- étudier l’effet d’une modification de la structure de mortalité sur la valeur de la garantie.


## 3. GMXB avec option de rachat (surrender)

Dossier : `PROJ/LEVY/GMXB_Surrender/`.
- `PROJ_GMXB_Surrender.m`, `Price_GMDB_BSM_NoSurrender.m` et `Scripts/Script_GMXB_Surrender_Table_Accuracy.m`.
- Combine une composante de mortalité, un compte investi (modèle de Lévy) et une option de rachat (exercice américain/Bermudan).
Travail :
- suivre le script pour reproduire les tables d’exactitude ;
- identifier la zone où la valeur de continuation bascule sous la valeur de rachat ;
- comparer avec les modules `Swing_Options`/`American_Options` pour réutiliser les routines de rollback.


## 4. Credit Default Swaps (CDS)

Dossier : `PROJ/LEVY/Credit_Default_Swaps/`.
- `PROJ_CDS.m` et `Script_Credit_Default_Swap.m` séparent jambes de prime vs protection.
- Intensité de défaut et recouvrement paramétrables.
Exercices :
- implémenter plusieurs profils d’intensité (constante, dépendante du sous‑jacent) ;
- vérifier la sensibilité aux hypothèses de recouvrement ;
- relier les résultats aux approches Monte Carlo ou PROJ classiques pour valider le pricing.
