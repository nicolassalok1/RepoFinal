# TP PROJ 6 – Monte Carlo, PDE/FD, arbres binomiaux et comparaisons

Ce TP recense les autres moteurs numériques fournis dans le dépôt pour valider les prix PROJ.


## 1. Librairie Monte Carlo (`Monte_Carlo/`)

Sous‑dossiers de payoffs : `American/`, `Asian/`, `Barrier/`, `Lookback/`, `ReturnBarrier/`, `ReturnTimer/`, `Variance_Swaps_Options/`…
Fonctions de simulation génériques :
- `Simulate_General_paths_func.m`, `Simulate_Jump_Diffusion_func.m`, `Simulate_Jump_Diffusion_Log_Returns_func.m` ;
- `Simulate_Heston_Euler_Schemes.m`, `Simulate_SLV_func.m`, `Simulate_RegimeSwitching_Diffusion_func.m`, `Simulate_RegimeSwitching_Diffusion_Unbiased.m` ;
- générateurs spécifiques (`DoubleExpoRnd.m`, `PoissonRnd.m`, `MixedNormalRnd.m`).
Travail :
- reproduire les scripts de prix Monte Carlo pour valider les résultats PROJ (même payoff, même modèle) ;
- benchmarker les estimateurs de variance (antithétique, control variate) ;
- instrumenter les simulateurs pour exporter des trajectoires (utiles dans les TPs 7–9).


## 2. PDE / Finite Difference

Dossier : `PDE_FiniteDifference/BlackScholes/`.
- `ExplicitFD_BlackScholes_func.m`, `ImplicitFD_BlackScholes_func.m`, `CrankNicolsonFD_BlackScholes_func.m`.
- Fournissent des solveurs 1D utilisables pour vérifier les prix PROJ sur grilles régulières.
Exercices :
- calibrer les triades (Nx, Nt, theta) pour éviter instabilités ;
- comparer les profils de valeur vs PROJ sur un call européen et une barrière.


## 3. Arbres binomiaux / trinômiaux et dossier `Comparisons/`

Dossier : `Lattice/BlackScholes/` + `Comparisons/`.
- `BinomialLattice_BlackScholes_func.m`, `TrinomialLattice_BlackScholes_func.m`, `Script_BlackScholes_BinomialLattice.m`.
- `Comparisons/Levy/Script_Compare_European.m`, `Comparisons/BlackScholes/`, `Comparisons/Heston/`, `Comparisons/SABR/`, `Comparisons/2D/`, `Comparisons/Regime_Switching/` compilent des tests croisés.
- Utilitaires : `Utils/binaryBracketIncreasing.m`, `Utils/calcBSImpVol.m`.
Travail :
- créer une table de comparaison (PDE vs Lattice vs PROJ) pour plusieurs maturités ;
- utiliser les scripts `Comparisons` pour valider plusieurs modèles en une seule passe ;
- repérer les cas où l’arbre est plus efficace que PROJ (petit nombre de pas / payoffs simples).
