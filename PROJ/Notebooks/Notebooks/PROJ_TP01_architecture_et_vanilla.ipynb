# TP PROJ 1 – Architecture PROJ et options vanille sous modèles de Lévy

Ce TP sert d’entrée en matière : comprendre comment `PROJ/LEVY/` est organisé et comment on construit un pricer vanille.
Objectifs :
- relier les briques génériques (`Helper_Functions/`, `RN_CHF/`, `Utils/`) aux scripts prêts à l’emploi ;
- identifier les paramètres clés de PROJ (taille de grille `N`, fenêtre `L1`, troncature `alpha`) ;
- préparer les TPs suivants sur les payoffs exotiques.


## 1. Construire l’entrée modèle PROJ

Dossiers : `PROJ/LEVY/Helper_Functions/` et `PROJ/LEVY/RN_CHF/`.
- `getModelInput.m`, `getModelInput_Heston.m` et `getTruncationAlpha.m` construisent la structure passée à tous les pricers PROJ (modèle, cf, discrétisation).
- `RN_CHF/cf_RN_*.m` et `SYMB_RN_*.m` regroupent les fonctions caractéristiques et symboles générateurs (BSM, Variance Gamma, CGMY, Kou, NIG, Bilateral Gamma, Heston...).
Travail :
- instancier plusieurs modèles via `getModelInput` et comparer les sorties (alpha, log‑spot shift, drift) ;
- vérifier comment les paramètres de volatilité/sauts se retrouvent dans `cf_RN_*` ;
- ajouter, si besoin, un modèle custom en copiant un template `SYMB_RN_*`.


## 2. Options européennes vanille (`European_Options/`)

Dossier : `PROJ/LEVY/European_Options/`.
- `PROJ_European.m` implémente la formule PROJ pour payoffs européens (call/put) sous modèle de Lévy.
- `Script_EuropeanOptions.m` illustre l’utilisation (choix du modèle, du payoff, de `N`, de `L1`).
À faire :
- lancer le script sur plusieurs modèles (BSM vs Variance Gamma) et mesurer l’impact de `N` et `alpha` ;
- comparer les prix PROJ au prix Black–Scholes analytique pour valider la calibration ;
- observer comment les grecs apparaissent via les dérivées du payoff tabulées dans le script.


## 3. Variantes vanille : forward‑start et surfaces de volatilité

Dossiers : `PROJ/LEVY/Forward_Starting_Options/` et `Utils/`.
- `PROJ_ForwardStarting.m` et `Script_ForwardStartingOption.m` ajoutent une phase d’observation avant l’activation du payoff.
- Les utilitaires `Utils/calcBSImpVol.m` et `Utils/binaryBracketIncreasing.m` permettent d’inverser la formule de Black–Scholes pour tracer des surfaces d’IV.
Travail suggéré :
- modifier le script forward‑start pour tester différentes dates de reset et vérifier l’impact sur le facteur d’actualisation ;
- réutiliser `calcBSImpVol` pour convertir les prix PROJ en vol implicite et vérifier la cohérence vis‑à‑vis des modèles analytiques ;
- documenter les paramètres à loguer systématiquement (N, alpha, payoff) pour les autres TPs.
