# TP PROJ 7 – Options multi‑actifs, méthode COS 2D et CTMC

Ce TP couvre les briques **multi‑actifs** non présentes dans `fypy/` : pricing 2D/3D par COS ou CTMC pour spreads, échanges et paniers.


## 1. Options d’échange et de spread : Analytical + COS 2D

Dossiers : `Analytical/BlackScholes/` et `Fourier/COS/`.
- `Price_Exchange_Option_Margrabe_2D.m` (Margrabe) et `Price_Spread_Option_BjerksundStensland_2D.m` / `Price_Spread_Option_Kirk_2D.m` donnent des références fermées.
- `Fourier/COS/Price_Exchange_COS2D.m` étend la méthode COS à un vecteur de log‑prix 2D corrélé.
Travail :
- comparer Margrabe vs COS en fonction de la corrélation $\\rho$ et du nombre de termes COS ;
- étudier les approximations de spread (Kirk/Bjerksund‑Stensland) vs COS pour différents ratios de strike ;
- implémenter une visualisation de la densité 2D tronquée utilisée par COS.


## 2. CTMC multi‑dimensionnel

Dossier : `CTMC/Diffusion_2D/` et `CTMC/Diffusion_3D/`.
- `price_2d_ctmc.m`, `price_3d_ctmc.m`, `get_transform_matrices_*`, `get_payoff_G_matrix_from_ygrid_*` construisent les grilles et matrices de transition.
- `Script_European_Barrier_Bermudan_Options.m` montre comment manipuler bermudanes/barrières sous diffusions corrélées.
À faire :
- dériver la matrice Q via `Q_Matrix.m` pour différentes corrélations et vérifier la stabilité ;
- comparer CTMC 2D vs COS 2D sur une option d’échange et un spread ;
- explorer la montée en dimension (3D) pour un panier worst‑of ou best‑of.


## 3. Exercices multi‑actifs et diagnostics

Ressources supplémentaires :
- `Comparisons/2D/` pour les tests croisés ;
- `Monte_Carlo/Simulate_Diffusion_2D.m` pour générer des trajectoires et vérifier les prix ;
- `Analytical/BlackScholes/BSM_Greeks.m` afin de récupérer les sensibilités par actif.
Travail :
- mettre en place un notebook de validation croisée (COS vs CTMC vs MC) ;
- analyser l’impact de la rotation de base (`decorrelate.m`) sur la rapidité de convergence ;
- documenter les limitations (taille de grille, mémoire) pour préparer les extensions 4D.
