# TP PROJ 9 – Swing, fader/range accrual et options timer

Ce TP regroupe les produits à **droits multiples** ou **accumulation de temps** : swing options, fader/range accrual, return/timer barriers.


## 1. Swing options (droits multiples)

Dossier : `PROJ/LEVY/Swing_Options/`.
- `PROJ_Swing_FixedRights.m`, `PROJ_Swing_LinearRec.m`, `PROJ_Swing_ConstantRecovery_Aug.m` + `Script_SwingOptions.m`.
- Sous‑dossier `coeff_funcs/` pour les fonctions de récupération/droits.
Travail :
- comparer les variantes (fixed rights vs linear/constant recovery) ;
- analyser l’arbre de décision (programmation dynamique) et les seuils d’exercice ;
- calibrer un cas de swing énergétique (plafond de droits) et observer la valeur marginale d’un droit supplémentaire.


## 2. Fader / Range accrual

Dossier : `PROJ/LEVY/Fader_Options/`.
- `Script_FaderOptions.m` décrit la structure range accrual (coupon activé lorsque le sous‑jacent reste dans une bande).
- Les paramètres `fader floor/cap`, présence d’une composante multiplicative, etc. y sont centralisés.
Exercices :
- tester plusieurs bandes de déclenchement et visualiser l’effet sur les coupons cumulés ;
- relier ce pricer aux step options du TP2 pour comprendre la continuité entre « soft barrier » et « range accrual » ;
- examiner comment PROJ gère le cumul du temps passé dans la bande.


## 3. Return Barriers et Range Return

Dossiers : `PROJ/LEVY/ReturnBarrier_Options/` et `Monte_Carlo/ReturnBarrier/`.
- `PROJ_ReturnBarrier.m`, `Script_ReturnBarrierOptions.m`, `getModel.m` : barrières basées sur le rendement cumulé.
- La version Monte Carlo permet de simuler le processus pour vérifier l’implémentation PROJ.
Travail :
- implémenter un scénario « range return » où la barrière porte sur $\\sum \\Delta S / S$ ;
- comparer PROJ vs MC pour diverses fenêtres d’observation ;
- étendre les scripts pour inclure des rebonds partiels ou des coupons conditionnels.


## 4. Return Timer

Dossier : `Monte_Carlo/ReturnTimer/`.
- Scripts MC dédiés aux options timer (maturité aléatoire déclenchée quand un cumul de rendement est atteint).
- Complément idéal au module PROJ (Parisian/Return Barrier) pour vérifier les lois de temps d’arrêt.
À faire :
- simuler plusieurs seuils de timer et tracer la distribution du temps d’arrêt ;
- brancher la sortie MC sur un pricing PROJ pour obtenir un « hybrid » (probabilité timer x payoff conditionnel) ;
- documenter les différences par rapport aux timers basés sur volatilité (non présents ici mais proches conceptuellement).
