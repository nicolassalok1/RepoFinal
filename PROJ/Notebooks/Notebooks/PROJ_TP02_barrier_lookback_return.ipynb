# TP PROJ 2 – Barrières, Parisiennes, Lookback et Step options

Ce TP couvre les payoffs liés à des niveaux de barrière dans `PROJ/LEVY/` : barrières simples/doubles, parisiennes, lookbacks, step (soft barrier).


## 1. Barrières simples et doubles

Dossier : `PROJ/LEVY/Barrier_Options/` et `PROJ/LEVY/ReturnBarrier_Options/`.
- `PROJ_Barrier.m`, `PROJ_Double_Barrier.m` et les scripts `Script_BarrierOptions.m`, `Script_DoubleBarrierOptions.m` couvrent les barrières up/down et les rebonds multiples.
- `ReturnBarrier_Options/` ajoute `PROJ_ReturnBarrier.m`, `getModel.m` et `Script_ReturnBarrierOptions.m` pour les barrières sur rendement cumulé.
Travail :
- reproduire les figures des scripts pour plusieurs combinaisons (in/out, rebated) ;
- analyser l’effet de la position de la barrière sur la fenêtre de troncature et sur la stabilité numérique ;
- connecter ces résultats avec les sections Monte Carlo (`Monte_Carlo/Barrier/`, `ReturnBarrier/`) pour validation croisée.


## 2. Parisian et timers

Dossier : `PROJ/LEVY/Parisian_Options/`.
- `PROJ_Parisian.m` et `Script_ParisianOptions.m` gèrent le comptage du temps passé sous/sur une barrière avant le déclenchement.
- Les timers (barrières retardées) réutilisent la même logique en combinant accumulation du temps et déclencheur discret.
Exercices :
- modifier le script pour comparer parisian cumulatif vs reset ;
- observer le rôle de la taille de pas (nombre d’observations) dans la précision ;
- relier les coefficients utilisés aux matrices de transition vues plus tard dans le TP9 (Return Timer).


## 3. Step / Soft barrier

Dossier : `PROJ/LEVY/Step_Options/`.
- `PROJ_StepOption_AutoParam.m` et `Script_StepOptions.m` implémentent les barrières « soft » (pénalité progressive).
- Le script ajuste automatiquement les paramètres recommandés (`N`, `alpha`) selon le profil de payoff.
À faire :
- jouer sur le paramètre de pente pour passer d’un step « dur » à un step quasi‑linéaire ;
- comparer les prix step avec ceux d’options barrières classiques pour vérifier la continuité ;
- noter les endroits où la sélection automatique de `alpha` n’est pas suffisante (cas extrêmes).


## 4. Lookback et hindsight options

Dossier : `PROJ/LEVY/Lookback_Options/`.
- `PROJ_Lookback.m` et `Script_Lookback_Options.m` traitent les payoffs basés sur les extrêmes (max/min) atteints par le sous‑jacent.
- Les formules nécessitent de stocker plusieurs grilles d’historiques (valeur courante + extrême).
Travail :
- tester la sensibilité du prix à la densité de grilles et au nombre d’états sauvegardés ;
- comparer les résultats PROJ avec une approximation type Monte Carlo lookback ;
- préparer une table récapitulative `barrière/soft/lookback` pour faciliter la sélection du pricer adapté.
