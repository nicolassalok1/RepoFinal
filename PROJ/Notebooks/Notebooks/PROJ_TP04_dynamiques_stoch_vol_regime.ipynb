# TP PROJ 4 – Dynamiques enrichies : stochastic volatility, régimes et temps changés

Ce TP étend PROJ au‑delà des modèles de Lévy purs : volatilité stochastique, régimes de Markov, processus temps‑changés et taux stochastiques.


## 1. Volatilité stochastique (`PROJ/STOCHASTIC_VOL/`)

Sous‑dossiers : `European/`, `Barrier/`, `American_Bermudan/`, `DV_Swaps_Options/`, `EIA_Cliquet/` + `Helper_Functions/`.
- Les scripts `Script_European.m`, `Script_European_Barrier_American.m`, etc. appellent des matrices générées par `get_SV_matrix_exponential.m`, `get_SV_matrix_expo_inputs.m`.
- Les fonctions `Q_Matrix.m` / `Q_Matrix_AllForms.m` construisent la discrétisation de la variance (Markov chain sur $v_t$).
Travail :
- dériver les matrices Q pour Heston vs 3/2 vs 4/2 et comparer les bornes `get_variance_grid_boundaries.m` ;
- tester les scripts européens/barrières sous différents régimes de corrélation $\\rho$ ;
- relier les prix obtenus aux approximations SABR/Hagan du TP10.


## 2. Régimes de Markov (`PROJ/REGIME_SWITCHING/`)

Dossier : `PROJ/REGIME_SWITCHING/European_Options/` + fonctions `get_RS_chf.m`, `get_RS_matrix_exponential.m`.
- `PROJ_RegimeSwitching_European.m` et `Script_EuropeanOptions.m` évaluent des options vanille où la dynamique des sauts/volatilité dépend d’un état de Markov.
- Les matrices de transition sont exponentiées pour obtenir des probabilités sur chaque pas.
À faire :
- calibrer un cas à deux régimes (calme/turbulent) et tracer la décomposition de prix par état ;
- modifier `get_RS_chf` pour introduire un drift dépendant de l’état ;
- comparer avec les librairies Monte Carlo `Simulate_RegimeSwitching_Diffusion_func.m`.


## 3. Processus temps changés (`PROJ/TIME_CHANGED/`)

Dossiers : `PROJ/TIME_CHANGED/European/`, `DV_Swaps_Options/`, `Helper_Functions/`.
- Les scripts utilisent une sous‑ordination (temps opérationnel) pour capturer la volatilité stochastique implicite.
- On retrouve des constructions analogues à celles du module Lévy, avec une composante additionnelle gérant la densité du temps changé.
Exercices :
- comparer les prix d’une option européenne temps changée vs Lévy standard avec mêmes moments ;
- étudier l’impact d’une sous‑ordination gamma sur les options de variance (`DV_Swaps_Options`) ;
- identifier comment réutiliser ces briques pour des payoffs plus complexes.


## 4. Taux stochastiques (`PROJ/STOCHASTIC_INTEREST/`)

Dossier : `PROJ/STOCHASTIC_INTEREST/One_Factor/` (sous‑dossiers `Bonds/`, `Equity_Swaps_Caps/`).
- `Price_Bond_Analytic_OneFactor.m` et `Price_Bond_CTMC_OneFactor.m` : pricing obligataire sous taux aléatoires.
- Les modules `Equity_Swaps_Caps` montrent comment coupler taux stochastiques et sous‑jacent actions.
Travail :
- comparer la solution analytique vs CTMC pour une obligation zéro‑coupon ;
- implémenter un pricing d’equity swap où PROJ traite le sous‑jacent et la partie taux est simulée par CTMC ;
- préparer les paramètres nécessaires pour combiner ces briques avec les contrats PROJ précédents.
