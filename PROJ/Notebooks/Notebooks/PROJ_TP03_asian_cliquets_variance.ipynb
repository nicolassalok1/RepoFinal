# TP PROJ 3 – Options asiatiques, cliquets et produits sur variance

Focus sur les payoffs dépendant de trajectoires/agrégats : moyennes (asiatiques), coupons resets (cliquets/EIA) et produits sur variance.


## 1. Asiatiques arithmétiques

Dossier : `PROJ/LEVY/Asian_Options/`.
- `PROJ_Asian.m` implémente l’approche PROJ pour les moyennes arithmétiques discrètes/continues.
- `Script_AsianOptions.m` montre comment choisir le nombre de dates d’observation et la grille de moyennes représentatives.
Travail :
- comparer les prix avec différentes granularités (mensuel vs quotidien) ;
- vérifier la convergence vis‑à‑vis des approximations analytiques (Vorst) disponibles dans `Approx/BlackScholes/Asian/` ;
- observer l’impact des sauts (Kou, VG) sur la distribution des moyennes.


## 2. Asiatiques géométriques et forward start

Dossiers : `PROJ/LEVY/Geometric_Asian_Options/` et `PROJ/LEVY/Forward_Starting_Options/`.
- `PROJ_Geometric_Asian.m` / `Script_GeometricAsianOptions.m` proposent un cas fermé pour vérifier l’implémentation.
- Le forward start (`PROJ_ForwardStarting.m`) sert de brique pour des contrats de type cliquet avec période d’accumulation.
Exercices :
- utiliser les géométriques comme benchmark pour valider la précision de l’algorithme arithmétique ;
- dériver un payoff « moyenne géométrique + reset » en combinant les scripts ;
- noter les différences de paramétrage (`L1`) entre géométrique (moins lourd) et arithmétique.


## 3. Cliquets et Equity Indexed Annuities

Dossier : `PROJ/LEVY/Cliquets/`.
- `PROJ_Cliquet.m`, `PROJ_MultiplicativeCliquet.m` et `Script_Cliquets_EIA.m` couvrent les cliquets additifs/multiplicatifs et les EIA.
- Les resets successifs appellent le pricer PROJ à chaque sous‑période via programmation dynamique.
Travail :
- reproduire les exemples du script (cap/floor, participation rate) ;
- ajouter une composante `multiplicative` pour simuler un produit equity‑linked ;
- analyser la sensibilité aux hypothèses de participation et aux spreads de taux.


## 4. Variance swaps et options sur variance

Dossier : `PROJ/LEVY/Variance_Swaps_Options/`.
- `PROJ_DiscreteVariance_Swaps_Options.m` + `Script_VarianceSwapsOptions.m` : variance swaps, variance/volatility options.
- Combine la projection PROJ avec les moments de log‑rendements pour valoriser les flux liés à $\\sum (\\Delta X_i)^2$.
À faire :
- mesurer l’influence du pas d’observation sur la variance réalisée ;
- comparer avec la librairie Monte Carlo (`Monte_Carlo/Variance_Swaps_Options/`) ;
- préparer les entrées nécessaires à la calibration des surfaces de variance implicite.


## 5. Bermudan/American sous modèle de Lévy

Dossier : `PROJ/LEVY/American_Options/`.
- `PROJ_Bermudan_Put.m` et `Script_BermudanOptions.m` résolvent les options à exercice anticipé via rollback PROJ.
- Supporte plusieurs modèles (BSM, Kou, VG…) à travers `getModelInput`.
Exercices :
- ajuster le nombre de dates d’exercice pour passer d’un Bermudan rare à quasi‑américain ;
- identifier les zones où la fonction valeur change rapidement (besoin d’une grille plus fine) ;
- rapprocher les résultats des approches spécifiques (CTMC ou MC American) pour valider la stabilité.
